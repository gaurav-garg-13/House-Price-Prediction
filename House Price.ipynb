# Feature Engineering

## Missing values in Categorical features

```python
f_na = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes == 'O']

for f in f_na:
    print(f, np.round(data[f].isnull().mean(),4),'% missing values')
```

```python
# Creating a new category in place of null for catergorical features

def replace_cat_features(data,features_nan):
    d = data.copy()
    
    d[features_nan] = d[features_nan].fillna('Missing')
    
    return d

data = replace_cat_features(data,f_na)
data[f_na].isnull().sum()
```

```python
data.head()
```

## Missing Values in numerical features

```python
f_na = [f for f in data if data[f].isnull().sum()>1 and data[f].dtypes!= 'O']
for f in f_na:
    print(f , np.round(data[f].isnull().mean(),4))
```

```python
# replaceig numerical null values

def replace_num_null(data,f_names):
    d = data.copy()
    for f in f_names:
        median_value = d[f].median()
        
        # Create a new feature that stores whether the feature had a null value there or not
        d[f + 'nan'] = np.where(d[f].isnull(),1,0)
        
        d[f].fillna(median_value,inplace=True)
    return d

data = replace_num_null(data,f_na)

data[f_na].isnull().sum()
```

```python
data.head()
```

### Temporal Variables (Date time Variables)

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature] = data['YrSold'] - data[feature]
```

```python
data.head()
```

```python
data[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

# Numerical Variables

Since Numerical variables were skewed we perform log normal distribution

```python
data.head()
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for f in num_features:
    data[f] = np.log(data[f])
```

```python
data.head()
```

# Handling Rare Categorical Features
Remove categorical variables that represent less than 1% of data

```python
cat_features = [f for f in data.columns if data[f].dtypes=='O']
```

```python
cat_features
```

```python
for f in cat_features:
    temp = data.groupby(f)['SalePrice'].count()/len(data)
    
    temp_df = temp[temp>0.01].index
    
    data[f] = np.where(data[f].isin(temp_df),data[f],'Rare_var')
```

```python
data.head(100)
```

```python
for feature in cat_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head(10)
```

## Feature Scaling

```python
scaled_features = [f for f in data.columns if f not in ['Id','SalePrice']]


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[scaled_features])
```

```python
X_train = pd.concat([data[['Id','SalePrice']].reset_index(drop =True),
                    pd.DataFrame(scaler.transform(data[scaled_features]), columns = scaled_features)],
                   axis = 1)
```

```python
X_train.to_csv('train.csv', index = False)
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
y_train = dataset[['SalePrice']]
```

```python
X_train = dataset.drop(['Id','SalePrice'],axis = 1)
```

```python
### Apply feature selection
#First specify lasso regression model and select suitable alpha

#Then use selectFormModel which select features which have non zero coefficients

feature_sel_model = SelectFromModel(Lasso(alpha = 0.005,random_state = 0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_features = X_train.columns[(feature_sel_model.get_support())]

print('Total Features: ', len(dataset.columns)-2)
print('Selected Features: ', len(selected_features))
print('Features with coefficients shrank to zero: ', np.sum(feature_sel_model.estimator_.coef_==0))
```
